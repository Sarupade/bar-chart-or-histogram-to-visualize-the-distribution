```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("task1.csv",skiprows=4)
```

```python
df_melted=df.melt(id_vars=['Country Name'],var_name='Year',value_name='Population')
```

```python
df_melted.dropna(subset=['Population'],inplace=True)
```

```python
top_10_countries = df[['Country Name', '2022']].sort_values(by='2022', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='2022', y='Country Name', data=top_10_countries, palette='coolwarm')
plt.xlabel('Population (Billions)')
plt.ylabel('Country')
plt.title('Top 10 Most Populated Countries (2022)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['2022'].dropna(), bins=30, kde=True, color='blue')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Histogram of Population Distribution (2022)')
plt.show()
```
